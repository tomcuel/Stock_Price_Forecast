```python
import yfinance as yf

ticker = "AAPL"
intervals = ["1m","5m","15m","30m","1h","1d"]

for itv in intervals:
    try:
        df = yf.download(ticker, interval=itv, period="max", progress=False)
        print(itv, "→", len(df), "rows")
    except:
        print(itv, "→ failed")
```
